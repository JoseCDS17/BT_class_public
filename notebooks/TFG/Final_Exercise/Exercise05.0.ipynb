```python
# import necessary modules
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np
import pandas as pd
from scipy.optimize import fsolve
from math import pi, exp
```

```python
#variables for different parameters
a=np.linspace(1e-9,1,1000)

w0 = [-1.2,-1.0, -0.8]
wa = [0.68,0.00, -0.77]
legarray = ['h = 0.6787', 'h = 0.6781', 'h = 0.6779']
```

```python
#Create figures
fig, ax = plt.subplots()

#plot the different values
for i in range(len(w0)):
    w = w0[i]+wa[i]*(1-a)
    if i == 0:
        ax.semilogx(a,w,color='b', linestyle='-')
    elif i == 1:
        ax.semilogx(a,w,color='r', linestyle='-')
    else:
        ax.semilogx(a,w,color='g', linestyle='-')

# output figure
ax.set_xlabel(r'$a$')
ax.set_ylabel(r'$\omega_\mathrm{fld}$')
ax.legend(legarray)
fig.tight_layout()
```

```python
#Create figures
fig_r, ax_r = plt.subplots()

#omega values for different densities
omega_r = 9.4e-5
omega_m = 0.32
omega_lambda = 0.68

#energy density evolution
for i in range(len(w0)):
    rho = omega_lambda*np.power(a,-3*(1+w0[i]+wa[i]))*np.exp(-3*wa[i]*(1-a))
    if i == 0:
        ax_r.loglog(a,rho,color='b', linestyle='-')
    elif i == 1:
        ax_r.loglog(a,rho,color='r', linestyle='-')
    else:
        ax_r.loglog(a,rho,color='g', linestyle='-')

legarray = ['h = 0.6787', 'h = 0.6781', 'h = 0.6779']

#radiation evolution
rho_r = omega_r*np.power(a,-4)
ax_r.loglog(a,rho_r,color='y', linestyle='-')
legarray.append(r'$\rho_r$')

#matter evolution
rho_m = omega_m*np.power(a,-3)
ax_r.loglog(a,rho_m,color='m', linestyle='-')
legarray.append(r'$\rho_m$')

# output figure
ax_r.set_xlabel(r'$a$')
ax_r.set_ylabel(r'$\rho / \rho_\mathrm{crit,0}$')
ax_r.set_xlim([1e-5,1])
ax_r.set_ylim([1e-8,1e15])
ax_r.legend(legarray)
fig_r.tight_layout()
```
